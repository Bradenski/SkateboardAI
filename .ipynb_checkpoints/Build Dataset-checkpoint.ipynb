```python
import os 
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def get_clip_info(clip: str) -> list:
    '''
    Returns info on clip. Assumes order of trick_firstname-lastname_sw_slow_mo
    ''' 
    sw = False
    slowmo = False

    clip = clip.split('/')[-1].split('\\')[-1].split('.')[0].split('_')
    trick = clip[0]
    skater = clip[1]
    if len(clip) > 2: 
        if clip[2] == 'sw':
            sw = True
        if clip[2] == 'slow-mo': 
            slowmo = True
        elif len(clip) > 3:
            if clip[3] == 'slow-mo':
                slowmo = True
            

    return [trick, skater, sw, slowmo]
```

```python
tricks_path = "C:/Users/Braden/Projects/SkateboardAI/skateboard_dataset/"
tricks_data = []

for clip in glob.glob(tricks_path + '/*'): 
    entry = get_clip_info(clip)
    entry.append(clip)
    tricks_data.append(entry)      
```

```python
trick_names = ['360-flip', 'crooked-grind' 'bs-flip', 'kickflip','ollie']

tricks_df = pd.DataFrame(tricks_data)
tricks_df = tricks_df.rename(columns={0: 'trick', 1: 'skater', 2: 'switch', 3: 'slowmo', 4: 'filepath'})
# tricks_df
categories = tricks_df['trick'].value_counts().index
counts = tricks_df['trick'].value_counts().values
plt.title("Trick data")
plt.bar(categories, counts, width=0.5)
```

```python
lengths = []
video_info = []
for trick_path in glob.glob(tricks_path+'/*'):
    cap = cv2.VideoCapture(trick_path)
    if cap.isOpened():
        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        fps = cap.get(cv2.CAP_PROP_FPS)
        duration = frame_count / fps if fps > 0 else 0
        
        lengths.append(frame_count)
        filename = os.path.basename(video_path)
        trick = filename.split('_')[0]
        
        video_info.append({
            'filename': trick_path,
            'trick': get_clip_info(trick_path)[0],
            'frames': frame_count,
            'duration': duration,
            'slow-mo': get_clip_info(trick_path)[-1]
        })
    cap.release()
# Show distribution
plt.figure(figsize=(10, 6))
plt.hist(lengths, bins=20, alpha=0.7)
plt.xlabel('Number of Frames')
plt.ylabel('Number of Videos')
plt.title('Video Length Distribution')
plt.grid(True, alpha=0.3)
plt.show()

# Show per-trick breakdown
tricks = {}
for info in video_info:
    trick = info['trick']
#     slow_mo = info['slow-mo']
    if trick not in tricks:
            tricks[trick] = [[],[]]
    tricks[trick][0].append(info['frames'])
    tricks[trick][1].append(info['slow-mo'])

print("\nPer-trick statistics:")
slow_mo_count = 0
for trick, stats in tricks.items():
    frame_counts = stats[0]
    slow_mo = stats[1]
    
    slow_mo_count += sum(slow_mo)
    print(f"{trick}: {len(frame_counts)} videos, "
          f"frames {min(frame_counts)}-{max(frame_counts)}, "
          f"avg {sum(frame_counts)/len(frame_counts):.1f}")
    print("Number of slow motion video clips: " + str(sum(slow_mo)))

print(f"Total videos: {len(lengths)}")
print(f"Total slow motion vidoes: {slow_mo_count}", )
print(f"Frame count - Min: {min(lengths)}, Max: {max(lengths)}")
print(f"Frame count - Mean: {sum(lengths)/len(lengths):.1f}")
print(f"Frame count - Median: {sorted(lengths)[len(lengths)//2]}")
```

### SkateboardingDataset

```python
import torch
from skimage import io, transform 
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils
import cv2
import sys
```

```python
class SkateboardDataset(Dataset): 
    def __init__(self, data_dir, transform=None):
        # parse filenames, store video paths
        self.data_dir = data_dir
        self.transform = transform
        self.video_paths = glob.glob(os.path.join(data_dir, "*"))
        
    def __len__(self): 
        # Return number of videos in dataset
        return len(self.video_paths)
    
    def preprocess_frame(self, frame, target_size=224):
        """Crop and resize frame image"""
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        h, w = frame.shape[:2]
        crop_size = min(h,w)
        start_h = (h - crop_size) // 2
        start_w = (w - crop_size) // 2
        cropped_frame = frame[start_h:start_h+crop_size, start_w:start_w+crop_size]
        resized_frame = cv2.resize(cropped_frame, (224,224))

        return resized_frame
        
    def load_video_frames(self, video_path): 
        
        cap = cv2.VideoCapture(video_path)
        frames = []
        if not cap.isOpened(): 
#             print("Error: could not open video.")
            pass
        else: 
            while cap.isOpened(): 
                ret, frame = cap.read() 
                # if frame is read correctly ret is True 
                if not ret: 
#                     print("Can't receive frame (stream end?). Exiting...")
                    break 
                frame = self.preprocess_frame(frame)
                frames.append(frame)
        cap.release()
        return frames
    
    def __getitem__(self, idx): 
        # load video, sample frames, return (frames, label)
        if torch.is_tensor(idx): 
            idx = idx.tolist()
            
        video_path = video_paths[idx]
        frames = self.load_video_frames(video_path)
        
#         label = video_path.split('\\')[-1].split('_')[0]
        label = get_clip_info(video_path)[0]
        
        return torch.tensor(frames), label
```
